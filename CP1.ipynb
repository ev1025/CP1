# 사전지식

### 데이터 분석 방법론 : CRISP-DM (CRoss Industry Standard Process for Data Mining)

**1) Business Understanding**
- 비즈니스 배경 및 목적 (비즈니스가 처한 상황, 문제점, 사업목표, 조직구성 등)   
`ex) 대리점 매출저하로 인한 사업성장문제를 개선하기 위한 방법`
- 분석 목적 및 목표 설정 (데이터분석 이유, 목적, 효과)   
`ex) 비즈니스 목적이 매출을 증대시키는 것 이라면 분석 목적은 증대 방법을 탐색하는 모델 수립`


**2) Data Understanding**
- 초기데이터 수집 : 고객데이터, 상품데이터, 구매데이터, 로그데이터 등
- 데이터 기술(describe) : 확보데이터에 대한 설명 자료   
`ex) 테이블 별 데이터 수, 테이블 항목 데이터유형(categoric, iterator, integer, numeric, character)`

**3) Data Preprocessing**
- 데이터누락, 오류, 의미해석불가 데이터들 =>정제
- 정제된 데이터로 새로운 데이터를 생성하거나 통합
- 데이터에 대한 이해를 통해 데이터 종류, 건수, 분석기간, 분석기준을 확정   
> **Tip** : 활용가능한 모든 데이터를 분석하기보다 목적에 맞게 최적화하여 분석(F검정 후 t검정을 통과하는 정도면 충분)

**4) Modeling**
- 모델 선정 가이드
   - 데이터량이 충분한가?(신뢰도)
   - 데이터의 품질이 중요한 모델인가?
   - 데이터의 형태가 모델에 적합한가? (변환을 하면 되지만, 성능에 영향을 줄 수 있다)
- 테스트 방안 설계,수립
   - 모델링 평가지표와 basline을 정해두어야한다.  
ex) 지도학습의 오류율 / 비지도학습의 오차값평균, 중심점, 실루엣   
    

**5) Evaluation**
- 모델의 성능이 목표와 일치 하는지, 비즈니스에 어떤 의미와 효과가 있을지 정리
- 평가지표 분석 및 모델성능 최적화(반복)


**6) Deployment(배포)**
- 결과 발표 및 보고 단계
- 데이터와 비즈니스에 대한 이해, 전처리단계, 모델링기술, 결과와 시사점을 정리 
- 향후 자동화 구축방안과 모니터링 방안을 추가

#### KDD(Knowledge Discovery in Database) - 고전 분석방법론

- 기술과 데이터베이스를 중심으로 insight 발굴을 위한 절차와 단계
- BI(Business Intelligence)라고도 불리며 DBMS를 운영하는 조직에서 사용가능하다.   

**1) Selection(데이터 선택)**
 - Raw data에서 분석에 필요한 자료를 선택, 도메인 이해와 과제목표설정)

**2) Prerprocessing(데이터 전처리)**
- 데이터 셋의 노이즈, 이상치, 결측치를 정제      

**3) Transformation(데이터 변환)** 
- 분석 목적에 따라 Feature를 생성, 선택하고 데이터마이닝 할 수 있도록 데이터를 분리

**4) Data Mining**   
- 분석목적에 맞는 데이터마이닝 알고리즘을 선택하여 데이터마이닝 실행

**5) Interpretation/Evaluation(해석/평가)**
- 데이터마이닝 결과를 해석,평가하고 결과가 충족하지 않았을 땐 데이터 선택과 데이터마이닝 반복수행

## 문제를 쪼개어 분석 : MECE(Mutually Exclusive Collectively Exhaustive)

**문제해결을 위한 이슈 분석**

**1. 핵심질문을 정의한다**
   - 과거,현재,미래 상황(배경)과 현재 직면한 문제 및 장애요소 정의

**2. 핵심질문을 이슈로 바꾼다.**
   - 핵심질문을 이슈로 발전시켜서 실천적이고 의사결정에 도움이 되도록 구체화

**3. 이슈를 잘게 쪼갠다.(MECE)**
   - 이슈를 상위이슈와 하위이슈(인과관계나 해결책)의 관계로 구성된다.
   - 하위 이슈 상호간에는 중복과 누락이 없어야한다.   
   `ex) 사람을 남자, 여자로 나누면 중복이없고, 누락되는 사람이 없다`
   - 하위 이슈의 단계는 충분히 검증 가능한 데이터 수집이나 분석을 할 수 있는 수준까지 나눈다.

**4. 쪼갠 이슈를 해결할 가설을 세우고 근거를 파악한다.**
- 도출된 이슈에 대한 가설을 미리 상정하고 이를 입증하기 위한 분석을 진행해야한다.
**5. 작업계획을 수립한다.**

## 가설수립 - 연역적 추론과 귀납적 추론

### 연역적 추론(Inductive Reasoning)


- 미리 정해놓은 예상 결과를 검증하는 방법   
- 데이터와 산출된 정보는 이론을 증명하고 검증하는 **수동적 역할**
- 가설의 **예외적인 결과에 대해서 설명하기 어렵다**는 단점이 있다.   
   `ex) 가설 : 매일 댓글을 다는 사용자는 액티브유저이다`   
     `-> 매일 접속은 하지만 댓글만 안다는 유저는 설명불가`

- **연구단계** : 배경지식 -> 가설 -> 연구설계 -> 측정 -> 자료수집 -> 분석 -> 경험적 일반화
   - 직접추론 : P이면 Q이다 -> Q이면 P이다(대우명제)
   - 간접추론 : 모든 사람은 죽는다 -> 나는 사람이다 - > 나는 죽는다.   



### 귀납적추론(Deductive Reasoning)

- 어떤 현상을 관찰하여 특성을 모두 열거하고 측정하여 유의미한 결과를 일반화 하는 방법(패턴찾기)
- 데이터는 이론을 새롭게 발견하고 명료화 하는 **능동적 역할**
- 데이터 분석가의 직관이 개입될 여지가 적다.
- **인과관계의 오류가 발생할 수 있고 일반화가 어렵다.**   
`ex) 맹자도 죽었다, 공자도 죽었다. 그러므로 사람은 죽는다`   
`ex) 비둘기는 난다, 참새도 난다, 그러므로 모든 새는 난다 => 일반화 모순`
- **연구단계** : 경험적 관찰 -> 가설 -> 연구설계 -> 측정 -> 자료수집 -> 분석 -> 이론수립

> **연역적 추론이 귀납적 추론보다 효율성이 높다.**
- 경험주의적 사고를 바탕으로하기 때문에 가설 생성에 소요되는 시간이 적다.
- 하지만 최근에는 기술의 발달로 데이터 수집과 처리가 빨라지면서 많은 데이터를 필요로하는 귀납적추론의 단점이 해소되는 상황

### 데이터분석에서 연역 + 귀납 상호보완적으로 사용

- 도메인지식과 분석가의 직관을 이용한 연역적 추론
- 데이터 탐색을 통한 패턴/특성을 도출하는 귀납적 추론
- 두 분석방법을 통해 MECE(=분할정복)

## 전환률(Conversion Rate, CVR)

- 전환수 / 클릭수 *100 = 전환률
- 클릭은 유효클릭과 무효클릭으로 나뉘며 전환률을 구할 때 따로 구분하진 않음
- 정밀하게 하기 위해선 유효클릭만 사용해도 좋음
